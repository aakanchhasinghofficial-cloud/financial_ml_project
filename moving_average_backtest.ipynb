## Moving Average Backtesting System

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# 100 NSE stocks:
url = "https://archives.nseindia.com/content/equities/EQUITY_L.csv"
df = pd.read_csv(url)
symbols = (df["SYMBOL"].head(100)+ ".NS").tolist()

print("Total Stocks:", len(symbols))
print(symbols[:20])
```

```python
# downloading historical data:
all_data = {}

for stock in symbols:
    
    data = yf.download(stock, start='2022-01-01', end='2025-12-21', auto_adjust = 'True', progress ='False')

    if not data.empty:
       all_data[stock]=data
       print(f"Downloaded {stock}")
```

```python
# apply SMA strategy:

results = []

for stock, data in all_data.items():
    data["SMA50"] = (data["Close"].rolling(50).mean())
    data["SMA200"] = (data["Close"].rolling(200).mean())

    data["Signal"] = 0
    data.loc[data["SMA50"]>data["SMA200"],"Signal"] = 1

    data["Position"] = (data["Signal"].diff())

    data['Returns'] = (data['Close'].pct_change())

    data["Strategy_Returns"] = (data["Returns"]*data["Signal"].shift(1))

    data["Equity_Curve"] = (1 + data["Strategy_Returns"]).cumprod()

    strategy_return = (data["Equity_Curve"].iloc[-1]-1)*100

    results.append({"Stock": stock, "Return (%)": round(strategy_return, 2)})
```

```python
# result dataframes:

results_df = pd.DataFrame(results)

results_df = results_df.sort_values(by = "Return (%)", ascending= False)

print("Top 10 performing stocks: ")
print(results_df.head(10))
```

```python
# Visualization:

stock = results_df.iloc[0]["Stock"]
data = all_data[stock]

buy_signal = data[data["Position"]==1]

sell_signal = data[data["Position"]== -1]

plt.figure(figsize=(14, 7))

plt.plot(data["Close"],label="Close Price")

plt.plot(data["SMA50"],label="SMA50")

plt.plot(data["SMA200"],label="SMA200")


plt.scatter(buy_signal.index, buy_signal["Close"], marker="^",color="green",s=100,label="BUY")

plt.scatter(sell_signal.index,sell_signal["Close"],marker="v",color="red",s=100,label="SELL")

plt.title(f"{stock} SMA Crossover Strategy")

plt.xlabel("Date")

plt.ylabel("Price")

plt.legend()

plt.grid(True)

plt.show()


```

```python
#Equity Curve 

plt.figure(figsize=(14,7))

plt.plot(data["Equity_Curve"], label ="Strategy Equity Curve")
plt.title(f"{stock} Equity Curve")

plt.xlabel("Date")

plt.ylabel("Portfolio Growth")

plt.legend()

plt.grid(True)

plt.show()
```
