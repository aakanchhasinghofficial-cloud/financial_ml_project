## Backtesting Engine Enhancement

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# 100 NSE stocks:
url = "https://archives.nseindia.com/content/equities/EQUITY_L.csv"
df = pd.read_csv(url)
symbols = (df["SYMBOL"].sample(100)+ ".NS").tolist()

print("Total Stocks:", len(symbols))
print(symbols[:20])
```

```python
# downloading historical data:
all_data = {}

for stock in symbols:
    
    data = yf.download(stock, start='2022-01-01', end='2025-01-01', auto_adjust = 'True', progress ='False')

    if not data.empty:
       data.columns = data.columns.get_level_values(0)
       all_data[stock]=data
       print(f"Downloaded {stock}")
```

```python
#Backtesting:

results = []

for stock, data in all_data.items():

    #moving average:
    data["SMA20"] = (data["Close"].rolling(20).mean())
    data["SMA50"] = (data["Close"].rolling(50).mean())
    
    # Buy/Sell Signals:
    
    data["Signal"] = 0
    data.loc[data["SMA20"]>data["SMA50"],"Signal"] = 1
    
    #Daily Returns:
    data['Returns'] = (data['Close'].pct_change())

    # Strategy returns:
    data["Strategy_Returns"] = (data["Returns"]*data["Signal"].shift(1))

    # Buy and Hold returns:
    data["Buy_Hold_Returns"] = (data["Returns"])
    data.dropna(inplace=True)
    if data.empty:
        continue
        
    # Equity Curves:
    data["Strategy_Equity"] = (1 + data["Strategy_Returns"]).cumprod()
    data["Buy_Hold_Equity"] = (1 + data["Buy_Hold_Returns"]).cumprod()

    # total return:
    strategy_total_return = (data["Strategy_Equity"].iloc[-1] - 1) * 100

    buy_hold_total_return = (data["Buy_Hold_Equity"].iloc[-1] - 1) * 100

    # CAGR:
    years = len(data) / 252

    strategy_cagr = (data["Strategy_Equity"].iloc[-1]** (1 / years) - 1) * 100

    buy_hold_cagr = (data["Buy_Hold_Equity"].iloc[-1]** (1 / years) - 1) * 100

    # maximum drawdown:
    strategy_peak = (data["Strategy_Equity"].cummax())

    strategy_drawdown = (data["Strategy_Equity"]- strategy_peak) / strategy_peak

    strategy_max_drawdown = (strategy_drawdown.min()) * 100
    
    #buy and hold drawdown
    
    buy_hold_peak = (data["Buy_Hold_Equity"].cummax())

    buy_hold_drawdown = (data["Buy_Hold_Equity"]- buy_hold_peak) / buy_hold_peak

    buy_hold_max_drawdown = (buy_hold_drawdown.min()) * 100 
```

```python

# Store Result:
results.append({"Stock": stock,
"Strategy Return (%)": round(strategy_total_return, 2),
"Buy & Hold Return (%)": round(buy_hold_total_return, 2),
"Strategy CAGR (%)": round(strategy_cagr, 2),
"Buy & Hold CAGR (%)": round(buy_hold_cagr, 2),
"Strategy Max Drawdown (%)": round(strategy_max_drawdown, 2),
 "Buy & Hold Max Drawdown (%)": round(buy_hold_max_drawdown, 2)
    })

# Results dataframes
results_df = pd.DataFrame(results)

print("\n BACKTEST RESULTS \n")

print(results_df.head())
```

```python
#Strategy vs Buy-and-Hold Plot
sample_stock = list(all_data.keys())[0]

sample_data = all_data[sample_stock]

plt.figure(figsize=(14, 7))

plt.plot(
    sample_data.index,
    sample_data["Strategy_Equity"],
    label="Strategy Equity Curve"
)

plt.plot(
    sample_data.index,
    sample_data["Buy_Hold_Equity"],
    label="Buy & Hold Equity Curve"
)

plt.title(
    f"{sample_stock} : Strategy vs Buy & Hold"
)

plt.xlabel("Date")

plt.ylabel("Portfolio Growth")

plt.legend()

plt.grid(True)

plt.show()

# results:
results_df.to_csv(
    "backtest_results.csv",
    index=False
)

print("\nResults saved successfully!")
```

Weakness of SMA Crossover Strategy

1. Moving averages are lagging indicators because they depend on past prices.

2. The strategy generates false signals during sideways markets.

3. Frequent trend reversals can create whipsaw losses.

4. Entry signals occur after the trend has already started.

5. The strategy does not include stop-loss or risk management techniques.
